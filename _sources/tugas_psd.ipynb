Judul Proyek

Prediksi Produksi dan Harga Beras Menggunakan Metode LSTM

1.	Pendahuluan 

Pada pendahuluan berisi gambaran umum tentang topik yang akan dibahas. Pada pendahuluan menjelaskan latar belakang masalah serta tujuan dari proyek sains data. (perusahan apa secara spesfik dan jelas )

  1.1	Latar Belakang

      Beras merupakan komoditas strategis di Indonesia karena merupakan makanan pokok bagi mayoritas penduduk. Fluktuasi produksi dan harga beras sangat dipengaruhi oleh berbagai faktor, termasuk cuaca, teknologi pertanian, kebijakan pemerintah, dan pasar global. Ketidakpastian dalam produksi dan harga beras dapat menimbulkan dampak besar terhadap ekonomi dan ketahanan pangan nasional.

  Dengan kemajuan teknologi, metode berbasis data semakin banyak digunakan untuk melakukan prediksi terhadap harga dan produksi beras di masa mendatang. Salah satu metode yang efektif untuk memprediksi data yang bersifat sekuensial adalah Long Short-Term Memory (LSTM), sebuah jenis Recurrent Neural Network (RNN) yang dirancang untuk memproses dan memprediksi data berurutan dengan pola jangka panjang.


  Visi dari laporan ini adalah untuk membantu pemerintah dan pemangku kepentingan dalam memprediksi produksi dan harga beras secara akurat guna mendukung stabilitas ketahanan pangan. 

  Misi penelitian ini adalah untuk menerapkan LSTM dalam prediksi harga beras, yang dapat membantu perencanaan distribusi dan stabilitas harga di masa depan.


  Strategi Bisnis:

    •	Meningkatkan akurasi prediksi produksi dan harga beras dengan menggunakan metode deep learning (LSTM).

    •	Berkolaborasi dengan pemerintah dan sektor swasta untuk penerapan hasil prediksi dalam kebijakan pangan.

    •	Menggunakan data berkelanjutan untuk meningkatkan ketahanan pangan.


  Target Bisnis:

    •	Mencapai akurasi prediksi lebih dari 90% dalam harga dan produksi beras.

    •	Meningkatkan efisiensi perencanaan distribusi dan stabilitas harga beras.

    •	Mengurangi ketidakstabilan harga beras di Indonesia.


  Tujuan Bisnis:

    •	Menciptakan model prediksi berbasis LSTM yang dapat diimplementasikan oleh pembuat kebijakan.

    •	Menyediakan sistem prediksi yang membantu perencanaan produksi dan distribusi beras berdasarkan pola harga yang diprediksi.

    •	Memberikan informasi yang berguna untuk menjaga kestabilan harga dan mengurangi fluktuasi yang ekstrem.


  1.2	Tujuan:

    •	Mengembangkan Model LSTM untuk Prediksi Harga Beras: Laporan ini bertujuan untuk membangun model prediksi harga beras berbasis Long Short-Term Memory (LSTM) yang mampu menangkap pola sekuensial dari data historis. Dengan memanfaatkan LSTM, laporan ini diharapkan memberikan solusi prediktif yang lebih akurat dibandingkan metode tradisional.

    •	Analisis Fluktuasi Harga Beras: Menganalisis faktor-faktor yang mempengaruhi fluktuasi harga beras, seperti kualitas beras dan pola musiman (bulan). Laporan ini akan mengeksplorasi bagaimana LSTM dapat mempelajari pola tersebut untuk memprediksi harga di masa depan.

    •	Mengevaluasi Performansi Model: Mengevaluasi performa model LSTM dalam hal akurasi prediksi dan membandingkannya dengan metode lain seperti time series atau machine learning konvensional. Tujuan ini adalah untuk memastikan bahwa metode yang digunakan memberikan hasil yang optimal.

    •	Memberikan Rekomendasi untuk Kebijakan: Berdasarkan hasil prediksi, laporan ini akan memberikan rekomendasi kebijakan kepada pemangku kepentingan untuk membantu menjaga stabilitas harga beras dan meningkatkan efisiensi dalam distribusi.


  1.3	Rumusan Masalah

    Dalam upaya mencapai visi, misi, strategi, target, dan tujuan bisnisnya, PT AgriBeras Nusantara menghadapi beberapa masalah yang bisa dipecahkan melalui analisis data, antara lain:

      •	Bagaimana cara membangun model prediksi harga beras yang akurat menggunakan metode LSTM?
      LSTM digunakan untuk memprediksi data sekuensial seperti harga beras, karena kemampuannya dalam menangkap pola jangka panjang dari data historis. Rumusan ini berfokus pada bagaimana LSTM dapat diterapkan untuk memprediksi fluktuasi harga berdasarkan data tahun dan bulan sebelumnya.

      •	Faktor apa saja yang mempengaruhi fluktuasi harga beras?
      Fluktuasi harga beras dipengaruhi oleh berbagai faktor, seperti kualitas beras, bulan, dan kondisi pasar. Pertanyaan ini akan mengeksplorasi faktor-faktor utama yang memengaruhi perubahan harga dan bagaimana model LSTM dapat mempelajari hubungan antar faktor tersebut.

      •	Seberapa akurat model LSTM dalam memprediksi harga beras dibandingkan dengan metode lainnya?
      Rumusan ini akan membandingkan kinerja LSTM dengan metode prediksi lain, seperti ARIMA, untuk melihat mana yang paling efektif dalam memberikan prediksi akurat.
